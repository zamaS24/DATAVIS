```python
df_net = pd.read_csv('netflix_titles.csv')

# Rajouter une colonne count, pour faciler le count
df_net['count'] = 1

df_net.head()
```

```python
# Create a new DataFrame with only relevant columns
df_filtered = df_net[['release_year', 'type', 'count']]

# Group by release_year, type, and count occurrences
grouped = df_filtered.groupby(['release_year', 'type']).sum()

# Reset index to make the grouped data more readable
grouped_reset = grouped.reset_index()

```

```python
movies = df_net2[df_net2['type'] == 'Movie']
movies
```

```python
import re

def find_number(text):
    # Check if text is a string, otherwise return an empty string
    if isinstance(text, str):
        num = re.findall(r'[0-9]+', text)
        return " ".join(num)
    else:
        return ""

# Apply the find_number function to handle non-string values
movies['durations'] = movies['duration'].apply(lambda x: find_number(x))

# Filter out rows where 'durations' is empty
movies = movies[movies['durations'] != ""]

# Convert durations to integer
movies['durations'] = movies['durations'].astype(int)
```

```python
movies['duration_bin'] = pd.qcut(movies['durations'], q=10) 
```

```python
movies['duration_bin'].value_counts()
```

```python
movies
```

```python
bins = [0, 60, 90, 95, 120, 150, 300, 350]
labels = ['0 - 60', '60 - 90', '90 - 95', '95 - 120', '120 - 150', '150  - 300', 'above 300']

movies['duration_bin'] = pd.cut(movies['durations'], bins, labels=labels, right=False)
movies
```
